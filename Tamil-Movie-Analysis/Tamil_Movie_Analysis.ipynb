```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Tamil movies.csv")

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.columns
```

```python
top_directors = df['director_name'].value_counts().head(10)

top_directors.plot(kind='bar')

plt.title("Top Tamil Movie Directors")
plt.xlabel("Director")
plt.ylabel("Number of Movies")
plt.show()
```

```python
top_actors = df['actor_1_name'].value_counts().head(10)

top_actors.plot(kind='bar')

plt.title("Top Lead Actors in Tamil Movies")
plt.xlabel("Actor")
plt.ylabel("Number of Movies")
plt.show()
```

```python
all_genres = df['genres'].str.split(',').explode()

genre_counts = all_genres.value_counts().head(10)

genre_counts.plot(kind='bar')

plt.title("Most Common Genres in Tamil Movies")
plt.xlabel("Genre")
plt.ylabel("Number of Movies")
plt.show()
```

```python
df['actor_pair'] = df['actor_1_name'] + " & " + df['actor_2_name']

top_pairs = df['actor_pair'].value_counts().head(10)

top_pairs.plot(kind='bar')

plt.title("Popular Actor Pairs")
plt.show()
```
