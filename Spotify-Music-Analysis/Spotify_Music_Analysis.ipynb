```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/spotify-tracks-dataset.csv")

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
top_artists = df['artists'].value_counts().head(10)

top_artists.plot(kind='bar')
plt.title("Top Artists on Spotify")
plt.xlabel("Artist")
plt.ylabel("Number of Songs")
plt.show()
```

```python
top_genres = df['track_genre'].value_counts().head(10)

top_genres.plot(kind='bar')
plt.title("Top Music Genres")
plt.show()
```

```python
sns.histplot(df['popularity'], bins=20)
plt.title("Song Popularity Distribution")
plt.show()
```

```python
df["duration_minutes"] = df["duration_ms"] / 60000

sns.histplot(df["duration_minutes"], bins=20)
plt.title("Song Duration Distribution")
plt.show()
```

```python
top_songs = df.sort_values("popularity", ascending=False).head(10)

top_songs[['track_name','artists','popularity']]
```
